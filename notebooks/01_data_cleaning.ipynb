# 01 - Data Cleaning 

This notebook cleans the BreathWatch Air Quality and Health 2015 - 2024 dataset in data/raw/global_air_pollution_dataset.csv to prepare the data for analysis and visualization on isolated variables. 

```python
# Setup and imports 

import pandas as pd 
import numpy as np 

pollution_df = pd.read_csv('../data/raw/global_air_pollution_dataset.csv')
pollution_df.describe()

```

The data is currently orgainzed by RECORD_ID. However, to make patterns and trends more visible, we want to sort the data according to year and season (starting from 2014 all the way to 2024 in the order of winter, spring, summer, fall). 

```python
# Sort dataframe by date and season 
season_order = ['Winter', 'Spring', 'Summer', 'Fall']
pollution_df['Date'] = pd.to_datetime (pollution_df['Date'])
pollution_df['Season'] = pd.Categorical(pollution_df['Season'], categories=season_order, ordered=True)
pollution_df.sort_values(by=['Date','Season'], inplace=True)

# Eliminate Record_ID column -> previous sequential ordering
pollution_df.drop(columns=['Record_ID'], inplace=True)

pollution_df.head(20)
```

Now we want to eliminate the NaN row entries to ensure only full completed records remain. 

```python
# Check for NaN values across df
print(pollution_df.isna().any())

# Eliminate those rows with empty values 
pollution_df.dropna(inplace=True)
pollution_df.describe()
```

```python
# Load data back into CSV for other notebooks to access
pollution_df.to_csv('../data/clean/cleaned_data.csv', index=False)
```
